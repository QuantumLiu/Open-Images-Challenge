```python
import pandas as pd

d = pd.read_csv('frcnn.csv')
```

```python
for j in range(len(d['PredictionString'])):
    if type(d['PredictionString'][j]) is str:
        l = d['PredictionString'][j].split(' ')
        for i in range(int(len(l)/6)):
            temp = l[i*6+2]
            l[i*6+2] = l[i*6+3]
            l[i*6+3] = temp
            temp = l[i*6+4]
            l[i*6+4] = l[i*6+5]
            l[i*6+5] = temp
        d['PredictionString'][j] = ' '.join(l)
```

```python
import pandas as pd

cl500 = list(pd.read_csv('challenge-2018-class-descriptions-500.csv', header=None)[0])
d = pd.read_csv('frcnns.csv')
```

```python
for j in range(len(d['PredictionString'])):
    if type(d['PredictionString'][j]) is str:
        l = d['PredictionString'][j].split(' ')
        new = []
        for i in range(int(len(l)/6)):
            score = float(l[i*6+1])
            if score > 0.1:
                for k in l[i*6:i*6+6]:
                    new.append(k)
        d['PredictionString'][j] = ' '.join(new)
```

```python
import pandas as pd

d = pd.read_csv('frcnn.new.csv')
cl500 = list(pd.read_csv('challenge-2018-class-descriptions-500.csv', header=None)[0])
for j in range(len(d['ImageId'])):
    if type(d['PredictionString'][j]) is str:
        l = d['PredictionString'][j].split(' ')
        new = []
        for i in range(int(len(l)/6)):
            score = float(l[i*6+1])
            classes = l[i*6]
            if score > 0.002 and classes in cl500:
                for h in l[i*6:i*6+6]:
                    new.append(h)
        d['PredictionString'][j] = ' '.join(new)
d.to_csv('frcnn.new.500.csv', mode='w', index=False)
```
