```python
for i in range(68388,len(filenames)):
    image = filenames[i]
    im = Image.open(os.path.join(image_path, image))
    w = im.width
    h = im.height
    detections = detector.detectObjectsFromImage(input_image=os.path.join(image_path, image), output_image_path="image_with_box.png", minimum_percentage_probability = 75)        
    pred_str = ""
    for eachObject in detections:    
        if eachObject["name"] in rev:
            b = np.array(eachObject["box_points"])/np.array([w,h,w,h])
            pred_str += ' ' + rev[eachObject["name"]] + ' %f'% (eachObject["percentage_probability"]/100) + ' %f %f %f %f' % tuple(b)
    ImageId.append(image.split('\\')[1].split('.')[0])
    PredictionString.append(pred_str)
    if i%100==0:
        print(i)
```
